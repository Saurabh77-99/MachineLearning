Linear Regression

    Y = w X + b

Y -> Dependent variable

X -> Independent variable

w -> weight (slope)

b -> bias (intercept)


weights and bias are model parameters

Learning rate and no of iterations are hyper-parameters



---


Gradient Descent

It is an optimization algorithm used for maintaining the loss function in various machine learning algorithms. It is used for updating the parameters of learning model.


    w = w - alpha * dw

    b = b - alpha * db

alpha is learning rate

Loss function is distance between your predicted value and your true value



---


Learning Rate:

It is a tuning parameter in an optimization algorithm that determines the each step size at each iteration while moving towards a minimum of loss function

In this we step down the curve to reach global minimum

dw -> differentiate cost function wrt slope

db -> differentiate cost function wrt intercept
