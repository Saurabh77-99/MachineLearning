Seaborn

Data Visualisation Library

```python
#importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has same builtin datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for your plots
sns.set_theme()
```

```python
# visualize th data
sns.relplot(data=tips,x='total_bill',y="tip",col="time",hue="smoker",style="smoker",size="size")
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.relplot(data=iris,x="sepal_length",y="petal_length",col = "species",hue="species")
```

Scatter Plot

```python
sns.scatterplot(x="sepal_length",y="petal_length",hue="species",data=iris)
```

```python
sns.scatterplot(x="sepal_length",y="petal_width",hue="species",data=iris)
```

TItanic Datasets

```python
#loading Titanic dataset
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

Count Plot

```python
titanic.shape
```

```python
sns.countplot(x="class",data=titanic)
```

```python
sns.countplot(x="survived",data=titanic)
```

Bar Chart

```python
sns.barplot(x="sex",y="survived",hue="class",data=titanic)
```

```python
# house price dataset
from sklearn.datasets import fetch_openml
housing = fetch_openml(name="house_prices", as_frame=True)
```

```python
house = pd.DataFrame(housing.data,columns=housing.feature_names)
house['PRICE'] =housing.target
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Correlation **bold text**

```python
correlation = titanic.corr()
```

```python
#constructing heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap="Blues")
```
