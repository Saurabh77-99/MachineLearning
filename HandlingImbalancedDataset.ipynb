Imbalanced dataset:

A dataset with an unequal class distribution



```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# determining distribution of two classes
credit_card_data['Class'].value_counts()
```

it is Highly imbalanced dataset and it has no missing values

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# separating the legit and fraud
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Under-Sampling method to handle imbalanced dataset

Building sample dataset containing single distribution of legit and fraud transactions

No. of fraud trans = 3

```python
legit_sample = legit.sample(n=3)
```

```python
print(legit_sample.shape)
```

Concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.tail()
```
